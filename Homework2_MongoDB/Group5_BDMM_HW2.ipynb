```python
import json
from bson.decimal128 import Decimal128
from datetime import datetime

# EXTRA: Save the document to a JSON file (to see the structure more clearly)
def convert_to_serializable(obj):
    """
    Converts non-serializable objects (like Decimal128 and datetime) to strings.
    """
    if isinstance(obj, Decimal128):
        return str(obj)
    elif isinstance(obj, datetime):
        return obj.isoformat()
    raise TypeError(f"Object of type '{obj.__class__.__name__}' is not JSON serializable")

def save_to_json_file(document, filename="airbnb_document.json"):
    """
    Saves a MongoDB document to a JSON file, handling non-serializable data types.
    """
    try:
        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(document, f, ensure_ascii=False, indent=4, default=convert_to_serializable)
        print(f"Document saved to {filename}")
    except Exception as e:
        print(f"Error saving to JSON file: {e}")
```

```python
# Function to recursively extract all fields from a document (nested dictionaries and lists)
def extract_fields(obj, prefix=""):
    """Recursively extract all fields from a document.

    Args:
        obj: The object to process (dict, list, or other type). 
        prefix (str, optional): String prefix for nested fields (e.g., "address" becomes "address.country"). Defaults to "".

    Returns:
        fields: A set of unique field names found in the document.
    """
    
    # Initialize an empty set to store unique fields
    fields = set()
    
    # Check if the object is a dictionary (e.g., a MongoDB document or subdocument)
    if isinstance(obj, dict):
        # Iterate over field:value pairs in the dictionary
        for k, v in obj.items():
            # Construct the full field name; no dot if prefix is empty, otherwise add dot separator
            full_field = f"{prefix}{k}" if not prefix else f"{prefix}.{k}"
            fields.add(full_field)                                                      # Add the full field to the set
            fields.update(extract_fields(v, full_field))                                # Recursively extract fields from the value, using the current field as prefix
            
    # Check if the object is a list (e.g., an array like 'reviews' or 'amenities')
    elif isinstance(obj, list):
        # Iterate over each item in the list
        for item in obj:
            # Recursively extract fields from the item, keeping the same prefix
            fields.update(extract_fields(item, prefix))
            
    return fields

# Aggregate to get all documents and process them
all_fields = set()
for doc in db.listingsAndReviews_HW2.find():
    doc_fields = extract_fields(doc)
    all_fields.update(doc_fields)

# Convert to sorted list for readability
unique_fields = sorted(list(all_fields))

# Print the list of unique fields
pprint(unique_fields)
print(f"In total, there are \033[1m{len(unique_fields)} unique fields\033[0m in the collection 'listingsAndReviews'")
```

#### **💬 1.1 | Confirm if All **`Reviews`** are in the same collection**

```python
# Identify if all copies of reviews contain the same data as 'reviews'
def compare_reviews_per_document(verbose=True):
    """
    Compares the 'reviews' array with all 'reviews_copy*' arrays for each document in the collection.
    Checks if 'reviews' contains all reviews from the copies (by content) on a per-document basis.

    Args:
        verbose (bool): If True, prints detailed results for each document as they are processed.

    Returns:
        dict: Summary of the comparison, including counts of mismatches and execution time.

    Notes:
        - Dynamically detects all fields matching 'reviews_copy*' using $objectToArray.
        - Compares full review content (e.g., _id, date, comments) using $setEquals.
        - Retains counts for additional insight.
        - Uses tqdm for progress tracking.

    Sources:
        - $group: https://www.mongodb.com/docs/manual/reference/operator/aggregation/group/
        - $project: https://www.mongodb.com/docs/manual/reference/operator/aggregation/project/
        - $unwind: https://www.mongodb.com/docs/manual/reference/operator/aggregation/unwind/
        - $objectToArray: https://www.mongodb.com/docs/manual/reference/operator/aggregation/objectToArray/
        - $setEquals: https://www.mongodb.com/docs/manual/reference/operator/aggregation/setEquals/
        - $filter: https://www.mongodb.com/docs/manual/reference/operator/aggregation/filter/
        - $map: https://www.mongodb.com/docs/manual/reference/operator/aggregation/map/
        - $reduce: https://www.mongodb.com/docs/manual/reference/operator/aggregation/reduce/
        - $concatArrays: https://www.mongodb.com/docs/manual/reference/operator/aggregation/concatArrays/
        - $regexMatch: https://www.mongodb.com/docs/manual/reference/operator/aggregation/regexMatch/
        - $size: https://www.mongodb.com/docs/manual/reference/operator/aggregation/size/
        - $ifNull: https://www.mongodb.com/docs/manual/reference/operator/aggregation/ifNull/
    """
    # Start timing the entire process
    start_time = time.time()

    # Count total documents for tqdm progress bar
    total_docs = db.listingsAndReviews_HW2.count_documents({})

    # Pipeline to process each document individually
    pipeline = [
        
        # Step 1: Group by _id to process each document separately
        {"$group": {
            "_id": "$_id",
            "all_fields": {"$first": "$$ROOT"}  # Keep the entire document
        }},
        
        # Step 2: Convert document fields to an array to find reviews_copy* dynamically
        {"$project": {
            "_id": 1,
            "reviews": {"$ifNull": ["$all_fields.reviews", []]},
            "copy_fields": {
                "$filter": {                                                       # Filter fields that start with 'reviews_copy'
                    "input": {"$objectToArray": "$all_fields"},
                    "cond": {"$regexMatch": {"input": "$$this.k", "regex": "^reviews_copy"}}
                }
            }
        }},
        
        # Step 3: Extract and compare content
        {"$project": {
            "reviews_content": "$reviews",                                          # Full content of reviews
            "copy_contents": {
                "$map": {                                                           # Extract the value (array) from each reviews_copy* field
                    "input": "$copy_fields", "in": {"$ifNull": ["$$this.v", []]}
                }
            },
            "reviews_count": {"$size": {"$ifNull": ["$reviews", []]}},
            "copy_counts": {
                "$map": {                                                           # Count reviews in each copy
                    "input": "$copy_fields",
                    "in": {"name": "$$this.k", "count": {"$size": {"$ifNull": ["$$this.v", []]}}
                    }
                }
            }
        }},
        
        # Step 4: Compare content and check containment 
        {"$project": {
            "all_copies_match_reviews": {
                "$reduce": {                                                         # Iterate over all copy arrays to compare with reviews
                    "input": "$copy_contents",
                    "initialValue": True,
                    "in": {
                        "$and": [
                            "$$value",                                               # Previous comparison result
                            {"$setEquals": [                                         # Compare full content
                                "$reviews_content",
                                {"$concatArrays": ["$reviews_content", "$$this"]}
                            ]}
                        ]
                    }
                }
            },
            "reviews_count": 1,                                                     # Return/Project counts for additional insight
            "copy_counts": 1
        }}
    ]

    # Execute pipeline and iterate with tqdm
    mismatches = 0                                              # Initialize mismatch counter
    cursor = db.listingsAndReviews_HW2.aggregate(pipeline)      # Execute the aggregation pipeline

    for doc in tqdm(cursor, desc="Comparing reviews per document"):
        doc_id = doc["_id"]
        all_match = doc["all_copies_match_reviews"]
        reviews_count = doc["reviews_count"]
        copy_counts = {item["name"]: item["count"] for item in doc["copy_counts"]}

        if not all_match:
            mismatches += 1
            if verbose:
                print(f"\nMismatch found in document _id: {doc_id}")
                print(f"Review counts - reviews: {reviews_count}, copies: {copy_counts}")
                print("Content in 'reviews' does not fully match all 'reviews_copy*' arrays.")
        elif verbose:
            print(f"\nDocument _id: {doc_id} - All 'reviews_copy*' arrays match 'reviews'.")
            print(f"Review counts - reviews: {reviews_count}, copies: {copy_counts}")


    # Calculate total execution time
    execution_time = time.time() - start_time

    # Summary of results
    summary = {
        "total_documents": total_docs,
        "mismatches_found": mismatches,
        "execution_time_seconds": execution_time
    }
    return summary

# Run the function
result = compare_reviews_per_document(verbose=False)
print("\nSummary:")
pprint(result)
```

```python
# Step 2: Aggregate to extract all fields and their types
pipeline = [
    {"$project": {"fields": {"$objectToArray": "$$ROOT"}}},
    {"$unwind": "$fields"},
    {"$project": {
        "field_name": "$fields.k",
        "field_value": "$fields.v",
        "type": {"$type": "$fields.v"}
    }}
]
raw_fields = list(db.listingsAndReviews_HW2.aggregate(pipeline))

# Step 3: Recursively flatten and collect types
def flatten_with_types(data, prefix=""):
    """Recursively flatten nested documents and track field paths with types."""
    field_types = {}
    if isinstance(data, dict):
        for key, value in data.items():
            full_key = f"{prefix}{key}" if prefix == "" else f"{prefix}.{key}"
            if isinstance(value, (dict, list)):
                sub_fields = flatten_with_types(value, full_key)
                field_types.update(sub_fields)
            else:
                field_types[full_key] = type(value).__name__
    elif isinstance(data, list):
        for i, item in enumerate(data):
            full_key = f"{prefix}[{i}]"
            if isinstance(item, (dict, list)):
                sub_fields = flatten_with_types(item, full_key)
                field_types.update(sub_fields)
            else:
                field_types[full_key] = type(item).__name__
    return field_types

all_field_types = {}
for doc in raw_fields:
    field_name = doc["field_name"]
    field_value = doc["field_value"]
    field_type = doc["type"]

    if field_type in ["object", "array"]:
        if field_name not in all_field_types:
            all_field_types[field_name] = set()
        all_field_types[field_name].add(field_type)
        flattened = flatten_with_types(field_value, field_name)
        for full_key, f_type in flattened.items():
            if full_key not in all_field_types:
                all_field_types[full_key] = set()
            all_field_types[full_key].add(f_type)
    else:
        if field_name not in all_field_types:
            all_field_types[field_name] = set()
        all_field_types[field_name].add(field_type)

# Step 4: Create DataFrame to make it easier to read and analyze
# Ensure all unique fields are included, even if not in all_field_types
df_data = {field: list(all_field_types.get(field, {""})) for field in unique_fields}
df = pd.DataFrame.from_dict(df_data, orient="index", columns=["datatypes"])
df.index.name = "field"

# Execution time
execution_time = time.time() - start_time
print(f"\nExecution time: {execution_time:.4f} seconds")
print(f"\nFound \033[1m{len(df)} unique fields\033[0m. DataFrame created with data types.\n")
df
```

```python
# Check 'n' and '%' of missing values for all fields in the collection and subdocuments

# Step 1: Aggregate to extract all fields and their values
pipeline = [
    {"$project": {"fields": {"$objectToArray": "$$ROOT"}}},
    {"$unwind": "$fields"},
    {"$project": {
        "field_name": "$fields.k",
        "field_value": "$fields.v",
        "type": {"$type": "$fields.v"}
    }}
]
raw_fields = list(db.listingsAndReviews_HW2.aggregate(pipeline))

# Step 2: Recursively flatten to get all field paths
def flatten_fields(data, prefix=""):
    """Recursively extract all field paths, including subdocuments and arrays."""
    field_paths = set()
    if isinstance(data, dict):
        for key, value in data.items():
            full_key = f"{prefix}{key}" if prefix == "" else f"{prefix}.{key}"
            field_paths.add(full_key)
            if isinstance(value, (dict, list)):
                field_paths.update(flatten_fields(value, full_key))
    elif isinstance(data, list):
        for i, item in enumerate(data):
            full_key = f"{prefix}[{i}]"
            field_paths.add(full_key)
            if isinstance(item, (dict, list)):
                field_paths.update(flatten_fields(item, full_key))
    return field_paths

all_field_paths = set()
for doc in raw_fields:
    field_name = doc["field_name"]
    field_value = doc["field_value"]
    field_type = doc["type"]
    all_field_paths.add(field_name)
    
    # Exclude specific fields from further processing - Because these documents exist in different quantities in different documents
    exclude = ["transactions", "amenities", "reviews", "host_verifications"]
    if field_type in ["object", "array"] and field_name not in exclude:
        if field_name in ["address.location.coordinates"]:
            field_value = field_value.get("coordinates", [])
        all_field_paths.update(flatten_fields(field_value, field_name))

unique_fields = sorted(list(all_field_paths))
# Drop "address.location.coordinates[0]" and "address.location.coordinates[1]" as they are part of "address.location.coordinates"
unique_fields.remove("address.location.coordinates[0]")
unique_fields.remove("address.location.coordinates[1]")
print(f"Found {len(unique_fields)} unique field paths.")

# Step 3: Count missing values for each field
missing_counts = {}
for field in unique_fields:
    # Count documents where field is missing, null, or empty string
    missing = db.listingsAndReviews_HW2.count_documents({
        "$or": [
            {field: {"$exists": False}},
            {field: None},
            {field: ""}
        ]
    })
    missing_counts[field] = missing

# Step 4: Create DataFrame with n and % missing
df_data = {
    "n": [missing_counts[field] for field in unique_fields],
    "%": [round((missing_counts[field] / total_docs) * 100, 2) for field in unique_fields]
}
df = pd.DataFrame(df_data, index=unique_fields)
df.index.name = "field"
df.sort_values(by="n", ascending=False, inplace=True)
df
```

```python
# Clean up - Remove fields with "" or None from individual documents
fields_to_clean = [field for field in unique_fields if missing_counts[field] > 0]
for field in fields_to_clean:
    # Remove field where it's "" or None
    db.listingsAndReviews_HW2.update_many(
        {"$or": [{field: ""}, {field: None}]},
        {"$unset": {field: ""}}
    )
    print(f"Cleaned '{field}' from {missing_counts[field]} documents.")
```

## **🏗 Data Model Adjustments**
**Goal**: Optimize the structure of the database by separating large arrays into their own collections while maintaining relationships through indexed fields.

### **📌 Key Changes**
1. **Created a new primary collection** → `Listings`
   - This collection stores only the most important fields required for a **customer-facing query**.
   - Large and growing arrays like `reviews`, `transactions`, and `host` were moved to separate collections to improve performance.

2. **Moved `reviews`, `transactions`, and `host` into separate collections**  
   - **Why?** Storing large arrays inside a single document causes slow queries and inefficient indexing.
   - **Solution:** We now store these as separate collections and link them via `LISTING_ID`.

3. **Indexed important fields for faster lookups**  
   - Indexes were added on fields like `LISTING_ID` and `HOST_ID` to **speed up queries**.

---

## **🔹 New Collections & Schema**

### **1️⃣ `Listings` (Primary Collection)**
Stores general listing information **without reviews, transactions, or full host details**.

---

### **2️⃣ `Hosts`**
Stores host information separately and links to `LISTINGS` using `HOST_ID`.
> 🔹 **Reasoning**: Hosts are often associated with multiple listings. Storing them separately avoids redundant data.

---

### **3️⃣ `Reviews`**
Stores all reviews separately and links to `LISTINGS` using `LISTING_ID`.

> 🔹 **Reasoning**:  
> - The number of reviews **grows continuously**. Keeping them inside `LISTINGS` would slow down reads.
> - A separate `REVIEWS` collection allows us to **easily paginate and analyze reviews over time**.

---

### **4️⃣ `Transactions`**
Stores past transactions separately and links to `LISTINGS` using `LISTING_ID`.

> 🔹 **Reasoning**:  
> - Customers do **not** need transaction history when browsing listings.
> - Business Intelligence teams **do need this data**, so we store it separately.

---

## **📌 Summary of Changes & Justification**

| **Change** | **Reason** | **Expected Improvement** |
|------------|------------|-------------------------|
| **Created `Listings` collection with only essential fields** | Eliminates unnecessary data in customer-facing queries | Faster query response times (~50% improvement) |
| **Moved `Reviews` to a separate collection** | Reviews grow indefinitely, slowing down queries | Easier to paginate reviews + optimized storage |
| **Moved `Transactions` to a separate collection** | Customers don’t need past transactions | Faster queries for customer pages |
| **Created `Hosts` collection** | Avoids redundancy (hosts can own multiple listings) | Reduces duplicate data and improves efficiency |
